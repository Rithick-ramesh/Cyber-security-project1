```
plt.pie(df['protocol_type'].value_counts(),
        labels=df['protocol_type'].value_counts().index,
        autopct='%1.1f%%')

plt.title("Protocol Types Used")
plt.show()
```

```
counts = df['protocol_type'].value_counts()

plt.bar(counts.index, counts.values)
plt.title("Protocol Types Used")
plt.xlabel("Protocol Type")
plt.ylabel("Count")
plt.show()
```

```
counts = df['protocol_type'].value_counts()

plt.plot(counts.index, counts.values, marker='o')
plt.title("Protocol Types Used")
plt.xlabel("Protocol Type")
plt.ylabel("Count")
plt.grid(True)
plt.show()
```

```
import matplotlib.pyplot as plt


counts = df['protocol_type'].value_counts()
labels = counts.index
values = counts.values


plt.figure(figsize=(12, 10))


plt.subplot(2, 2, 1)
plt.pie(values, labels=labels, autopct='%1.1f%%')
plt.title("Protocol Types Used")

plt.subplot(2, 2, 2)
plt.bar(labels, values)
plt.title("Protocol Types Used")
plt.xlabel("Protocol Type")
plt.ylabel("Count")


plt.subplot(2, 2, 3)
plt.plot(labels, values, marker='o')
plt.title("Protocol Types Used")
plt.xlabel("Protocol Type")
plt.ylabel("Count")
plt.grid(True)


plt.tight_layout()
plt.savefig("protocol_dashboard.png", dpi=300)
plt.show()

```
